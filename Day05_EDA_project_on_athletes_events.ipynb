```python
data.groupby('Team')['Medal'].count().sort_values(ascending=False)
```

```python
data.groupby(by=['Year','Team'])['Medal'].count()
```

```python
#What is the average age,height,weight of athletes by age,height,weight?

data.groupby(by=['Sex','Sport'])[['Age','Height','Weight']].mean().reset_index()
```

```python
#identify the athletes who have participated in multiple olympic games
multiple_participation=data.groupby(by=['Name','Year'])['Medal'].count().reset_index()
multiple_participation[multiple_participation['Medal']>1]
```

```python
#What are the most common heights and weights of medal winning athletes?
data[data.Medal.notna()][['Height','Weight']].value_counts().reset_index().sort_values(by='Weight')
```

```python
#how has number of athletes participating evolved over the years
data.groupby('Year')['Name'].count().plot()
```

```python
data.groupby('Year')['Name'].count().plot(kind='hist')
```

```python
data.groupby('Year')['Name'].count().plot(kind='pie')
```

```python
#what is the gender distribuition of athletes in different sports over the years?
data.groupby(['Year','Sport','Sex'])['Name'].value_counts().reset_index()
```
